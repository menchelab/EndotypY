#### or just load the finished example object from within package data

```python
import pickle
import pkgutil

#load prepared exapmple
data = pkgutil.get_data('EndotypY', 'data/example_obj.pkl')
pex = pickle.loads(data)
```

```python
# read in a graph from a file
path_ppi = 'ppi_tiny.txt'
endo.import_network(path_ppi)

path_seed = 'example_seeds.txt'
#read in a seed set from a file
endo.import_seeds(path_seed)
```

```python
#prepare RWR
endo.prepare_rwr(r=0.8)
```

```python
#run the RWR
endo.run_rwr(scaling=True, k=200)
```

```python
nx.draw(endo.connected_subgraph, with_labels=True, node_color='lightblue',
        edge_color='gray', node_size=200, font_size=6,
        font_color='black')
```
